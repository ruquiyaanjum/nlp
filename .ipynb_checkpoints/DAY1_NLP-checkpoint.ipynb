text preprocessing :
  1) data cleaning : $ Food is good @#  --> Food is good {[A-Za-z]} (coding)
  2) lowering :  food is good
                 11   24 55
                 
  ex: machine translation --> english -- spanish [A-Za-z0-9]

Tokenization: splitting sentences into smaller units

ex: food is good. i will not come to eat.
1) sentence tokenization (paragraph to sentence)  -- food is good(1). I will not come to eat(2)
2) word tokenization (sentence to words) -- food(1) is(2) good(3).

why tokenization is needed?
    easy for computation and easy for processing
    1000 paragraph to 100 sentences (using tokenizations) 5 to 6 words.
